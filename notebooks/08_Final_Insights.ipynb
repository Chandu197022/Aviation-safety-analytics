```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/clean_incidents.csv")
```

```python
print(f"Total Incidents: {len(df)}")
print(f"Total Fatalities: {df['fatalities'].sum()}")
print(f"Total Injuries: {df['injuries'].sum()}")
print(f"Fatal Incidents: {df['is_fatal'].sum()}")
```

```python
top_airlines = (
    df["airline"]
    .value_counts()
    .head(10)
)

top_airlines.plot(kind="bar")
plt.title("Top Airlines by Incidents")
plt.show()
```

```python
weather_risk = (
    df.groupby("weather_conditions")["fatalities"]
    .sum()
    .sort_values(ascending=False)
)

weather_risk.head(10).plot(kind="bar")
plt.title("Fatalities by Weather")
plt.savefig(
    "../visuals/weather_risk.png",
    bbox_inches="tight")
plt.show()
```

```python
phase_risk = (
    df.groupby("phase_of_flight")["fatalities"]
    .sum()
    .sort_values(ascending=False)
)

phase_risk.plot(kind="bar")
plt.title("Fatalities by Flight Phase")
plt.savefig(
    "../visuals/phase_risk.png",
    bbox_inches="tight")
plt.show()
```

```python
manufacturer_risk = (
    df.groupby("manufacturer")["fatalities"]
    .sum()
    .sort_values(ascending=False)
)

manufacturer_risk.head(10).plot(kind="bar")
plt.title("Top Manufacturers by Fatalities")
plt.show()
```

```python
yearly = (
    df.groupby("year")
    .size()
    .reset_index(name="incidents")
)

sns.lineplot(data=yearly, x="year", y="incidents")
plt.title("Incidents Over Time")
plt.show()
```

```python
summary = pd.DataFrame({
    "Metric": [
        "Total Incidents",
        "Total Fatalities",
        "Total Injuries"
    ],
    "Value": [
        len(df),
        df["fatalities"].sum(),
        df["injuries"].sum()
    ]
})

summary
```
