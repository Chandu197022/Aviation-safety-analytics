```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(
    "../data/clean_incidents.csv"
)
```

```python
yearly = (
    df.groupby("year")
    .size()
    .reset_index(
        name="incidents"
    )
)
```

```python
plt.figure(
    figsize=(14,6)
)

sns.lineplot(
    data=yearly,
    x="year",
    y="incidents"
)

plt.title(
    "Yearly Aviation Incidents"
)
plt.savefig(
    "../visuals/yearly_trend.png",
    bbox_inches="tight"
)

plt.show()
```

```python
top_airlines = (
    df["airline"]
    .value_counts()
    .head(15)
)
```

```python
top_airlines.plot(kind="bar")

plt.title("Top Airlines")

plt.savefig(
    "../visuals/top_airlines.png",
    bbox_inches="tight"
)

plt.show()
```
