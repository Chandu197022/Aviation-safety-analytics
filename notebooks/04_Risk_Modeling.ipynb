```python
import pandas as pd

df = pd.read_csv(
    "../data/clean_incidents.csv"
)
```

```python
risk = (
    df.groupby(
        "airline"
    )
    .agg({
        "fatalities":"sum",
        "incident_id":"count"
    })
)
```

```python
risk["risk_score"] = (
    risk["fatalities"]
    /
    risk["incident_id"]
)
```

```python
risk.sort_values(
    "risk_score",
    ascending=False
)
```
