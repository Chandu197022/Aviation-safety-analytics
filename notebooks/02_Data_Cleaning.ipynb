```python
import pandas as pd

incidents = pd.read_csv(
    "../data/incidents.csv"
)
```

```python
incidents.columns = (
    incidents.columns
    .str.lower()
    .str.strip()
)
```

```python
incidents.isnull().sum()
```

```python
num_cols = [
    "fatalities",
    "injuries",
    "occupants_onboard"
]

for col in num_cols:

    incidents[col] = (
        incidents[col]
        .fillna(0)
    )
```

```python
incidents.drop_duplicates(
    inplace=True
)
```

```python
incidents.to_csv(
    "../data/clean_incidents.csv",
    index=False
)
```
